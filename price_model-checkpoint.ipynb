```python
#Suppression des colonnes ayant plus de 90% de données manquantes
df=df.dropna(thresh=len(df)*0.90,axis=1)
```

```python
df.shape
```

```python
# Categorisation des colonnes(numeriques & object )
numerical_columns=[]
object_columns=[]
for col in df.columns:
   
    print(col,'-------->',df[col].unique().size,"  type=",df[col].dtypes)
    
        
    if df[col].dtypes != "object" and col!= "SalePrice":
        numerical_columns.append(col)
    elif col!="SalePrice":
        object_columns.append(col)
```

```python
X=df.drop(["SalePrice"],axis=1)
y=df["SalePrice"]
```

```python
# Creation des pipelines

from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import PolynomialFeatures


numerical_pipeline=make_pipeline(SimpleImputer(),StandardScaler(),PolynomialFeatures()) #polinomialfeature
object_pipeline=make_pipeline(SimpleImputer(strategy='most_frequent'),OneHotEncoder(sparse=False)) 
```

```python
preprocessor=make_column_transformer((numerical_pipeline,numerical_columns),(object_pipeline,object_columns))
```

```python
#Preprocessing & normalisation de données et mis a jour de df avec ces données.
a=preprocessor.fit_transform(X)
encoded_data=pd.DataFrame(a,columns=preprocessor.get_feature_names_out(X.columns))
df.drop(X.columns,axis=1,inplace=True)
df=pd.concat([df,encoded_data],axis=1)
```

```python
#Elimination des données peu correlées
corr_threshold=0.20
corr_with_target=df.corr()["SalePrice"].abs()
```

```python
columns_to_keep=corr_with_target[corr_with_target >= corr_threshold ].index
df=df[columns_to_keep]
```

```python
X=df.drop("SalePrice",axis=1)
y=df["SalePrice"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=0)

```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
print("Score du modele:",model.score(X_test,y_test))
#model.predict(X_test)
```
